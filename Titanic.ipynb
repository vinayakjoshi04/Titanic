```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Cleaning of Data Starts From Here

```python
titanic_train.dropna(subset=['Embarked'], inplace=True)
```

```python
x = titanic_train[['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']]
y = titanic_train['Survived']
```

```python
x.head()
```

```python
y.head()
```

```python
x.isnull().mean()*100
```

```python
x['Age'].fillna(x['Age'].median(), inplace=True)
```

```python
x.isnull().mean()*100
```

```python
x.sample(5)
```

```python
x['Cabin']
```

```python
x.sample(5)
```

```python
x.isnull().mean()*100
```

```python
x['FamilySize'] = x['SibSp'] + x['Parch']
```

```python
x.sample(5)
```

```python
x['IsAlone'] = 1  # default to alone
x.loc[x['FamilySize'] > 1, 'IsAlone'] = 0  # not alone if FamilySize > 1
```

```python
x.sample(5)
```

```python
x.drop(columns=['SibSp', 'Parch'], inplace=True)
```

```python
x.head()
```

```python
x['Title'] = x['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())
```

```python
x['Title'] = x['Title'].replace(['Mlle', 'Ms'], 'Miss')
x['Title'] = x['Title'].replace(['Mme', 'Lady', 'Countess', 'Dona'], 'Mrs')
x['Title'] = x['Title'].replace(['Dr', 'Rev', 'Col', 'Major', 'Capt', 'Sir', 'Don', 'Jonkheer'], 'Rare')
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['Title'] = le.fit_transform(x['Title'])
```

```python
x.head()
```

```python
x['NameLength'] = x['Name'].apply(len)
x.drop('Name', axis=1, inplace=True)
```

```python
x.head()
```

```python
x['Sex'] = x['Sex'].map({'male': 0, 'female': 1})
```

```python
x.head()
```

```python
x['Embarked'].value_counts()
```

```python
x['Embarked'] = x['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
x.head()
```

```python
x['Cabin'].value_counts()
```

```python
# Extract deck letter from Cabin
x['Deck'] = x['Cabin'].astype(str).str[0]

# Fill missing deck values with 'M' (for Missing)
x['Deck'] = x['Deck'].replace('n', 'M')  # if 'nan' becomes 'n'

# Encode Deck
from sklearn.preprocessing import LabelEncoder
deck_encoder = LabelEncoder()
x['Deck'] = deck_encoder.fit_transform(x['Deck'])

# Optional: drop original Cabin
x.drop('Cabin', axis=1, inplace=True)
```

```python
x['Deck']
```

```python
x.head()
```

```python
x.drop('Ticket', axis=1, inplace=True)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(max_iter=1000)
LR.fit(X_scaled, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(n_estimators=100, random_state=42)
RFC.fit(x, y)
```

```python
from xgboost import XGBClassifier
XGB = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
XGB.fit(x,y)
```
